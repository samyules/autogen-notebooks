```python
from google.colab import userdata
import os
from crewai import Agent, Task, Crew, Process
from langchain.chat_models import ChatOpenAI

os.environ['OPENAI_API_KEY'] = userdata.get('OpenAI_API')
```

```python
# Define your agents with roles and goals

boss = Agent(
    role='Boss',
    goal='Oversee a team of agents writing a blog post about machine learning',
    backstory="You are a world class researcher working at a major data science company, recently promoted to supervisor",
    verbose=True,
    allow_delegation=True,
    llm=ChatOpenAI(temperature=0.5, model_name="gpt-4-1106-preview")
)

editor = Agent(
    role="Editor",
    goal="Quality Assurance and Quality control of a blog post for a major data science company. Ensure the content is enganging and truthful.",
    backstory="You are a copy editor at a major data science company",
    verbose=True,
    allow_delegation=True,
    llm=ChatOpenAI(temperature=0.2, model_name="gpt-4-1106-preview")
)

keyword_researcher = Agent(
    role="Keyword Researcher",
    goal="Conduct keyword research",
    backstory="An SEO specialist with expertise in keyword research and search trends",
    verbose=True,
    allow_delegation=False,
    llm=ChatOpenAI(temperature=0.3, model_name="gpt-4-1106-preview")
)

outliner = Agent(
    role="Outliner",
    goal="Outline a blog post about machine learning",
    backstory="You are a great skilled content strategist who understands how to structure blog posts for maximum impact",
    verbose=True,
    allow_delegation=False,
    llm=ChatOpenAI(temperature=0.5, model_name="gpt-4-1106-preview")
)

writer = Agent(
    role="Writer",
    goal="Write an informative, engaging, SEO optimized blog post",
    backstory="You are a great technical writer at a major data science company, specialized on writing machine data related content",
    verbose=True,
    allow_delegation=False,
    llm=ChatOpenAI(temperature=0.9, model_name="gpt-4-1106-preview")
)

proofreader = Agent(
    role="Proofreader",
    goal="Ensure the blog post is free from grammatical and spelling errors",
    backstory="You are a meticulous proofreader with an eye for detail and a passion for perfect grammar",
    verbose=True,
    allow_delegation=False,
    llm=ChatOpenAI(temperature=0.2, model_name="gpt-4-1106-preview")
)
```

```python
# Create Tasks for your agents
task1 = Task(description='Oversee the process of writing a blog post on the latest in AI advancements', agent=boss)
task2 = Task(description='Once the blog post is written, ensure that it is engaging, informative, truthful, and SEO optimized', agent=editor)
```

```python
# Initialize your crew and a sequential process.
crew = Crew(
    agents=[boss, editor, keyword_researcher, outliner, writer, proofreader],
    tasks = [task1, task2],
    verbose=2,
    process=Process.sequential
)
```

```python
# Get your crew to work
result = crew.kickoff()
```
